```python
import pingouin as pg

url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/' +
       'Estadistica-machine-learning-python/master/data/births.csv')

datos = pd.read_csv(url,sep=',')
datos.head()
```

**Prueba T-Student**

Hipotesis nula

$H_0+\mu_1 = \mu_2$

Estadistico de prueba

$T=\frac{\bar{x}-\bar{y}+(\mu_1-\mu_2)}{\sqrt{\frac{S_1^2}{m} * \frac{S_2^2}{n}}}$

El estadistico de prueba tiene una distribución t-student

```python
#filto de la base de datos (fumadores, no-fumadores)
w_smokers = datos.loc[datos.smoke=='smoker','weight']
w_nonsmokers = datos.loc[datos.smoke=='nonsmoker','weight']
```

```python
#Aplicación del t-test
pg.ttest(x=w_smokers,y=w_nonsmokers)
```

Como p-valor=0.1379>0.05

Se encuentra dentro de la región de aceptación

Media de los cuadrados del error

$MSE=\frac{S_1^2+S_2^2+...+S_n^2}{I}$
